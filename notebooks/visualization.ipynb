```python
import py3Dmol
```

```python
path = "../data/AF-P76011-F1-model_v6.pdb"
```

```python
with open(path, "r") as f:
    pdb_data = f.read()
```

```python
view = py3Dmol.view()
```

```python
view.addModel(pdb_data, "pdb")
view.setStyle({"cartoon": {"color": "spectrum"}})
view.zoomTo()
view.show()
```
